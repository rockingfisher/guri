```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/bigreant.xlsx')
df.head(1)
```

```python
df.drop(['Unnamed: 0','번지','본번','부번','계약기간','계약구분','갱신요구권 사용','종전계약 보증금 (만원)','종전계약 월세 (만원)'], axis=1, inplace=True)
```

```python
df
```

```python
df['계약년월'] = pd.to_datetime(df['계약년월'], format='%Y%m')
df['단지명'] = df['단지명'].str.replace(r'\([^)]*\)', '', regex=True)
df['전용면적(㎡)'] = df['전용면적(㎡)'].round(-1)
df['ID'] = df['시군구'] + ' ' + df['단지명'] + ' ' + df['전용면적(㎡)'].astype('str')
```

```python
df['분기'], df['년도'] = (df['계약년월'].dt.quarter, df['계약년월'].dt.year)
df
```

```python
table1 = pd.pivot_table(df, values=['전용면적(㎡)','보증금(만원)','월세(만원)','건축년도'], index=['년도','분기'], columns=['ID'], aggfunc=np.mean)
table1
```

```python
# check null 
check_null = table1.isna().sum() / 18

# columns of null ratio >= 0.5
check_null[check_null >= 0.5]
# remove columns of null ratio >= 0.5
remove_cols = check_null[check_null >= 0.5].keys()
table1 = table1.drop(remove_cols, axis=1)

table1
```

```python
table1.fillna(method='ffill', inplace=True)
table1.fillna(method='bfill', inplace=True)
```

```python
table1.round(1)
```

```python
table1.to_excel('data/rent.xlsx')
```
