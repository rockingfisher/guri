```python
import pandas as pd
import numpy as np

df = pd.read_excel('data/인덱스있는버전(매매만).xlsx')
df.head()
```

```python
# 아파트 : 전용면적 기준, 
# area1) 40㎡이하, 
# area2) 40㎡초과~60㎡이하, 
# area3) 60㎡초과~85㎡이하, 
# area4) 85㎡초과~102㎡이하,
# area5) 102㎡초과~135㎡이하, 
# area6) 135㎡초과
```

```python
df.tail()
```

```python
# area1) 40㎡이하, 

area1 = df.loc[df['전용면적(㎡)'] <= 40]
area1
```

```python
# area2) 40㎡초과~60㎡이하, 

area2 = df['전용면적(㎡)'] > 40
area3 = df['전용면적(㎡)'] <= 60

# df.loc[area2 & area3]
area2 = df.loc[area2 & area3]
area2
```

```python
# area3) 60㎡초과~85㎡이하, 

area3 = df['전용면적(㎡)'] > 60
area4 = df['전용면적(㎡)'] <= 85

area3 = df.loc[area3 & area4]
area3
```

```python
# area4) 85㎡초과~102㎡이하,

area4 = df['전용면적(㎡)'] > 85
area5 = df['전용면적(㎡)'] <= 102

area4 = df.loc[area4 & area5]
area4
```

```python
# area5) 102㎡초과~135㎡이하, 

area5 = df['전용면적(㎡)'] > 102
area6 = df['전용면적(㎡)'] <= 135

area5 = df.loc[area5 & area6]
area5
```

```python
# area6) 135㎡초과

area6 = df.loc[df['전용면적(㎡)'] > 135]
area6
```

```python
print(len(area1))
print(len(area2))
print(len(area3))
print(len(area4))
print(len(area5))
print(len(area6))
```
