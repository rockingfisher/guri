```python
from haversine import haversine, Unit
from geopy.geocoders import Nominatim

import numpy as np
import pandas as pd
import os
import re
```

```python
ilho = pd.read_excel('data/1호선.xlsx', header=None)
del ilho[0]
address = ilho[2]
address
```

```python
ilho
```

```python
geo_local = Nominatim(user_agent='South Korea') # 지역선택

def geocoding(address): 
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y

latitude = []
longitude = []
exception_addr = []

for i in address:
    try:
        latitude.append(geocoding(i)[0])
        longitude.append(geocoding(i)[1])
    except:
        latitude.append(i)
        longitude.append(i)
        exception_addr.append(i)

address_df = pd.DataFrame({'역명':ilho[1], '위도':latitude,'경도':longitude})
address_df
```

```python
address_df.to_excel('data/ilho.xlsx')
```
