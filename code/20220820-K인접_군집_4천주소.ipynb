```python
import pandas as pd
import numpy as np

import folium
from haversine import haversine, Unit
from geopy.geocoders import Nominatim
from folium.plugins import MarkerCluster
```

```python
df_4000 = pd.read_excel('data/실거래가/03_결측치처리후_주소지_4천개.xlsx', index_col=0)

map = folium.Map(location=[37.545510798074794, 126.95256795474381], zoom_start=12)
marker_cluster = MarkerCluster().add_to(map)

for i in range(len(df_4000)):
    try:
        folium.CircleMarker([df_4000.iloc[i,1], df_4000.iloc[i,2]], radius=8, fill_color='yellow').add_to(marker_cluster)
    except:
        pass

map
```
