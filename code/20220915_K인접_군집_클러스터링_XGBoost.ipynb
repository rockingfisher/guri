```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/10_4천주소_10변수.xlsx')
df
```

```python
df.corr()
```

```python
from sklearn import metrics
from sklearn.datasets import load_boston

data = df
X = data.drop(['BrandPower', 'ID'], axis = 1)
y = data['BrandPower']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 11)

import xgboost
xgb = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)

xgb.fit(X_train, y_train)
y_preds = xgb.predict(X_test)


print('R^2:',metrics.r2_score(y_test, y_preds))
print('MSE:',metrics.mean_squared_error(y_test, y_preds))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_preds)))
```
