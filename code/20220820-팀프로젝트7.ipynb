```python
from haversine import haversine, Unit
from geopy.geocoders import Nominatim

import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('data/줄인파일.xlsx')
df.drop(['Unnamed: 0','2018/1','2018/2','2018/3','2018/4','2019/1','2019/2','2019/3','2019/4','2020/1','2020/2','2020/3','2020/4','2021/1','2021/2','2021/3','2021/4','2022/1','2022/2'], axis=1, inplace=True)
df.head()
```

```python
df['address'] = df['시군구'] + ' ' + df['도로명']
df.head(5)
```

```python
address = df['address']

geo_local = Nominatim(user_agent='South Korea') # 지역선택

def geocoding(address): 
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y

latitude = []
longitude = []
exception_addr = []

for i in address:
    try:
        latitude.append(geocoding(i)[0])
        longitude.append(geocoding(i)[1])
    except:
        latitude.append(i)
        longitude.append(i)
        exception_addr.append(i)

address_df = pd.DataFrame({'ID':df['ID'], '위도':latitude,'경도':longitude})
address_df
```

```python
address_df.to_excel('data/주소지.xlsx')
```
