```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/ID있는버전.xlsx')
df
```

```python
df['계약년월'] = pd.to_datetime(df['계약년월'], format='%Y%m')
```

```python
df['분기'], df['년도'] = (df['계약년월'].dt.quarter, df['계약년월'].dt.year)
df
```

```python
table1 = pd.pivot_table(df, values=['전용면적(㎡)','거래금액(만원)','건축년도'], index=['년도','분기'], columns=['ID'], aggfunc=np.mean)
table2 = pd.pivot_table(df, values=['전용면적(㎡)','거래금액(만원)','건축년도'], index=['ID'], columns=['년도','분기'], aggfunc=np.mean)
table1
```

```python
# check null 
check_null = table1.isna().sum() / 18

# columns of null ratio >= 0.5
check_null[check_null >= 0.5]
# remove columns of null ratio >= 0.5
remove_cols = check_null[check_null >= 0.5].keys()
table1 = table1.drop(remove_cols, axis=1)

table1
```

```python
table1.fillna(method='ffill', inplace=True)
table1.isnull().sum()
```
