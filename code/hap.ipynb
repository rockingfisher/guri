```python
import pandas as pd
```

```python
df1 = pd.read_excel('data/아파트(매매)_실거래가18.xlsx')
df1
```

```python
df2 = pd.read_excel('data/아파트(매매)_실거래가18~19.xlsx')
df3 = pd.read_excel('data/아파트(매매)_실거래가19~20.xlsx')
df4 = pd.read_excel('data/아파트(매매)_실거래가20~21.xlsx')
df5 = pd.read_excel('data/아파트(매매)_실거래가21~22.xlsx')
```

```python
df_rent1 = pd.read_excel('data/아파트(전월세)_실거래가_18.xlsx')
df_rent2 = pd.read_excel('data/아파트(전월세)_실거래가_18~19.xlsx')
df_rent3 = pd.read_excel('data/아파트(전월세)_실거래가_19~20.xlsx')
df_rent4 = pd.read_excel('data/아파트(전월세)_실거래가_20~21.xlsx')
df_rent5 = pd.read_excel('data/아파트(전월세)_실거래가_21~22.xlsx')
```

```python
bigdf_sale = pd.concat([df1,df2,df3,df4,df5])
```

```python
bigdf_rent = pd.concat([df_rent1,df_rent2,df_rent3,df_rent4,df_rent5])
bigdf_rent.info()
```

```python
bigdf_rent.to_excel('data/bigreant.xlsx')
```

```python
bigdf_sale = bigdf_sale.sort_values('계약년월')
```

```python
bigdf_sale.drop(['번지','본번','부번','해제사유발생일','중개사소재지'], axis=1, inplace=True)
bigdf_sale
```

```python
bigdf_sale['전용면적(㎡)'] = bigdf_sale['전용면적(㎡)'].round(-1)
bigdf_sale
```

```python
bigdf_sale.drop(bigdf_sale[bigdf_sale['거래유형'].str.contains('직거래')].index, inplace=True)
bigdf_sale
```

```python
bigdf_sale['단지명'] = bigdf_sale['단지명'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
bigdf_sale.to_excel('data/아파트매매정리.xlsx')
```
